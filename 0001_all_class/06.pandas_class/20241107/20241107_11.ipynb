```python
import pandas as pd
df = pd.read_csv('students.csv', index_col='no')
df
```

students.xlsx 파일을 불러와서
평균 80점 이상, 국어 90점 이상을 출력하세요

```python
mask = (df['kor'] >= 90) & (df['avg'] >= 80)
df[mask]
```

평균이 60점 이상 또는 국어 70점 이상

```python
mask = (df['avg'] >= 60) | (df['kor'] >= 70)
df[mask]
```

```python
# str 문자열
df = pd.read_excel('score.xlsx')
mask = df['이름'].str.startswith('강')
# 가 포함되어 있는 이름을 검색 : contains
mask = df['이름'].str.contains('근')
df[mask]
df['SW특기'].str.lower()
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)
# isin() : 리스트에 있는 단어가 포함되어 있는지 체크
df[filt]

# 소문자 취환 : isin() 확인
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# java 라는 글자 있는지 체크
# na = True : NaN True인식, na=False : NaN False인식
filt = df['SW특기'].str.contains('Java')
df[filt]
```
