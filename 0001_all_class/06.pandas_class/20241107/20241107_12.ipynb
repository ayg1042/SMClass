```python
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx')
```

```python
# 결측치 데이터 (NaN)
# NaN데이터 채우기 : fillna()
df.fillna(' ', inplace=True)
df
```

```python
# 학교 컬럼에 NaN데이터 입력
# df['학교'] = np.nan
df.fillna('없음') # inplace=True 모든 NaN값에 해당데이터를 입력
# df['SW특기'].fillna('없음') # 해당 컬럼에만 적용
```

```python
# NaN 데이터 삭제 : dropna()
# NaN 데이터가 있는 row 삭제
df = pd.read_excel('score.xlsx', index_col='지원번호')
df.dropna(axis='index', how='any')
```

```python
df['학교'] = np.nan
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
# dropna : NaN 데이터 삭제, axis : 축 (index, columns)
# how default : any(NaN이 하나라도 있으면), all(전부 NaN일때)
df.dropna(axis='columns')
```

* fillna : NaN 데이터 해당값으로 채워준다.
* dropna : NaN 데이터를 삭제
- axis : index, columns
- how : 어떤경우에 삭제할지 결정(any - 한개라도 있으면, all - 모두 NaN일때 삭제)
- numpy.nan : NaN값 생성
