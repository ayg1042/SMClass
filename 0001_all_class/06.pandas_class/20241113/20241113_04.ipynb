```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
print('{:.2f}%'.format(10))
num = 10.2
print(f'{num}%')
print(f'{num:.2f}%')
```

```python
values = [30, 25, 20, 13, 10, 2]
explode = [0.1, 0.2, 0, 0, 0, 0]
# wedgeprops = 중간공백
wedgeprops = {'width': 0.6}
labels = ['python', 'java', 'javascript', 'oracle', 'html', 'etc']
color = ['plum', 'peachpuff', 'c', 'bisque', 'khaki', 'ivory']
plt.pie(values, labels=labels, explode=explode, counterclock=False, startangle=90, autopct='%.1f%%', colors=color, wedgeprops=wedgeprops)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
explode = [0.1, 0.2, 0, 0, 0, 0]
# wedgeprops = 중간공백
# 'edgecolor':'black'
# 'linewidth': 2
wedgeprops = {'width': 0.6, 'edgecolor':'black', 'linewidth': 2}
labels = ['python', 'java', 'javascript', 'oracle', 'html', 'etc']
color = ['plum', 'peachpuff', 'c', 'bisque', 'khaki', 'ivory']

# 데이터 값 : 2% 이하의 값은 출력되지 않도록 함수 적용하세요.
# pct : values의 퍼센트 값 생성
# def custom_autopct(pct):
#   if pct >= 10:
#     # return ('%.1f%%' % pct)
#     # return ('{:.1f}%'.format(pct))
#     return (f'{pct:.1f}%')
#   else:
#     return ''

def custom_autopct(pct):  
  return f'{pct:.1f}%' if pct >= 10 else f'{pct:.0f}%'


# pctdistance = 글자위치
plt.pie(values, labels=labels, counterclock=False, startangle=90, autopct=custom_autopct, colors=color, wedgeprops=wedgeprops, pctdistance=0.7)
plt.legend(loc=(1.2, 0.3))
plt.show()
```
