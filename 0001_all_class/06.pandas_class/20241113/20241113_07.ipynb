```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
# 영화 관객수를 막대그래프로 출력하세요.
# 평균 막대그래프를 추가하고, 평균 선을 그어서 표시하세요.
# 이름 : 평균, 년도 2024, 관객수 : 평균관객수, 평점 : 평균 평점
# df['평균'] = df['관객 수'].mean()
df
```

```python
df.loc[8] = ['평균', '2024', df['관객 수'].mean(), df['평점'].mean()]
```

```python
plt.figure(figsize=(15, 5))
test = plt.bar(df['영화'], df['관객 수'], width=0.3)
test[len(df['관객 수'])-1].set_color('red')
plt.plot(df['관객 수'].index, [df.loc[8]['관객 수']] * 9, ls='--', color='red')
plt.text(df['영화'][0], df['관객 수'][0], int(df['관객 수'][0]), ha='center')
plt.xticks(rotation=90)
plt.ylim(1000, 1800)
plt.show()
```
