```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('출산율2.xlsx', skiprows=2,nrows=2, index_col=0)
df
```

```python
df.index = ['출생아 수', '합계 출산율']
df.info()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df.columns, df.iloc[0], marker='o', color='red', label='출생아 수')
plt.plot(df.columns, df.iloc[1], marker='o', color='green', label='합계 출산율')
plt.legend()
plt.show()
```
