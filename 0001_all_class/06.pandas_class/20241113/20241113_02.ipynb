```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
def cm_func(x):
  return str(x) + 'cm'

df['키'] = df['키'].apply(cm_func)
df['키']
```

```python
def k_func(x):
  if x < 50:
    return 50
  return x

# 50점 아래 값을 50점으로 리턴해서 출력하세요.
# int, float, str 단일값은 함수명으로 호출
for i in range(len(df.index)):
  print(k_func(df['국어'][i]))
```

```python
def text(x):
  return int(x/2)

# 누적 막대그래프를 사용해서 값들을 모두 중앙에 출력하세요.
w = 0.25
plt.bar(df['이름'], df['국어'], width=w)
plt.bar(df['이름'], df['영어'], bottom = df['국어'], width=w)
plt.bar(df['이름'], df['수학'], bottom = df['국어'] + df['영어'], width=w)
for i in range(len(df.index)):
  plt.text(df['이름'][i], text(df['국어'][i]), df['국어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i] + text(df['영어'][i]), df['영어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i] + text(df['수학'][i]), df['수학'][i], ha='center')
plt.show()
```
