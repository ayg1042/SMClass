```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('연령별인구현황01.xlsx', skiprows=3, nrows=1, usecols='B, E:Y', index_col='행정기관')
df2 = pd.read_excel('연령별인구현황01.xlsx', skiprows=3, nrows=1, usecols='B, AB:AV', index_col='행정기관')
```

```python
df2.iloc[0] = df2.iloc[0].str.replace(',', '').astype(int)
```

```python
gx, gy = df.columns, df2.iloc[0]/1000
plt.barh(gx, gy)
```

```python
df.iloc[0] = df.iloc[0].str.replace(',', '').astype(int)
```
