```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
plt.title('학생 성적 그래프')
plt.plot(df.index, df['국어'], label='국어')
plt.plot(df.index, df['영어'], label='영어')
plt.plot(df.index, df['수학'], label='수학')

for i in range(len(df.index)):
  plt.text(df.index[i], df['국어'][i], df['국어'][i])
  plt.text(df.index[i], df['영어'][i], df['영어'][i])
  plt.text(df.index[i], df['수학'][i], df['수학'][i])

plt.legend()
plt.show()
```

```python
def func(x):
  if x - 30 < 0:
    return x
  else:
    return x-30
```

```python
# 누적 막대그래프
plt.figure(figsize=(8, 4))
plt.bar(df['이름'], df['국어'], width=0.6, label='국어')
plt.bar(df['이름'], df['영어'], bottom=df['국어'], width=0.6, label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어'] + df['영어'], width=0.6, label='수학')
plt.title('학생 성적 그래프')
plt.legend(ncol=3)

for i in range(len(df['이름'])):
  plt.text(df['이름'][i], func(df['국어'][i]), df['국어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i], df['영어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i] + df['수학'][i], df['수학'][i], ha='center')

plt.show()
```
