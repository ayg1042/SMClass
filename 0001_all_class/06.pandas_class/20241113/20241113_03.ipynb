```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

```python
values = [30, 25, 20, 13, 10, 2]
explode = [0.1, 0.2, 0, 0, 0, 0]
labels = ['python', 'java', 'javascript', 'oracle', 'html', 'etc']
color = ['plum', 'peachpuff', 'c', 'bisque', 'khaki', 'ivory']
plt.title('프로그램 사용률')
# pie(데이터값) labels : 글자 출력, explode : 원 그래프 분리, counterclock : 시계반대방향, False 시계방향
# startangle = 90 : 12시 부터 시작함, autopct='%.1f%%' : 퍼센트 출력
plt.pie(values, labels=labels, explode=explode, counterclock=False, startangle=90, autopct='%.1f%%', colors=color)
plt.legend(loc=(1.2, 0.3))
plt.show()
```
