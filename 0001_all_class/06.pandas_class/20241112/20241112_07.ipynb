```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8] #sm 매출액
y2 = [5, 1, 3] # lg 매출액
y3 = [1, 2, 5] # sk 매출액


plt.figure(figsize=(10, 5))
plt.plot(x, y, label='SM 매출액', marker='o', ls='-.')
plt.plot(x, y2, label='LG 매출액', marker='x', ls='-')
plt.plot(x, y3, label='SK 매출액', marker='v', ls='--')

# 그래프 내 값출력
for i in range(len(x)):
  plt.text(x[i], y[i] + 0.2, y[i], ha='center')
  plt.text(x[i], y2[i] + 0.2, y2[i], ha='center')
  plt.text(x[i], y3[i] + 0.2, y3[i], ha='center')

plt.title('기업별 매출액')
# 기본 : 세로 출력
# ncol = 가로
plt.legend(ncol=3)
plt.show()

```
