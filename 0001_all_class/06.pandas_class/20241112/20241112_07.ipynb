```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8] #sm 
y2 = [5, 1, 3] # lg 
y3 = [1, 2, 5] # sk 


plt.figure(figsize=(10, 5))
plt.plot(x, y, label='SM ', marker='o', ls='-.')
plt.plot(x, y2, label='LG ', marker='x', ls='-')
plt.plot(x, y3, label='SK ', marker='v', ls='--')

#   
for i in range(len(x)):
  plt.text(x[i], y[i] + 0.2, y[i], ha='center')
  plt.text(x[i], y2[i] + 0.2, y2[i], ha='center')
  plt.text(x[i], y3[i] + 0.2, y3[i], ha='center')

plt.title(' ')
#  :  
# ncol = 
plt.legend(ncol=3)
plt.show()

```
