```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
label = ['홍길동', '유관순', '강감찬']
values = [180, 187, 184]

colors = ['r', 'g', 'b']

plt.figure(figsize=(5, 3))
plt.title('막대그래프')

# bar 그래프
# 1개의 색상으로 처리
# plt.bar(label, values, color='red')
# 여러 색상 사용
plt.bar(label, values, color=colors, width=0.2, label='학생키')

plt.xticks(label, ['1번', '2번', '3번'], rotation=45) # x라벨 변경
plt.yticks([0, 50, 100, 150, 200, 250, 300])
plt.legend(loc='upper center')

# 선 그래프
# plt.plot(label, values)
```

막대 그래프 무늬 설정

```python
colors = ['red', 'green', 'blue']
bar = plt.bar(label, values, width=0.5, color=colors, alpha=0.5)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
# 막대그래프의 무늬
```

* 범위설정 xticks, yticks = 0부터 표시된다.
* 특정 구간부터 범위 설정 : ylim

```python
colors = ['red', 'green', 'blue']
bar = plt.bar(label, values, width=0.5, color=colors, alpha=0.5)
# plt.yticks([100, 150, 200, 250]) # 0부터 시작
plt.ylim(176, 190) # 범위 지정
plt.xticks()
```

barh : 옆으로 누운 막대그래프

```python
label = ['홍길동', '유관순', '강감찬']
values = [180, 187, 184]
colors = ['r', 'g', 'b']
# 옆쪽으로 나온다.
plt.barh(label, values, color=colors, height=0.2)
```
