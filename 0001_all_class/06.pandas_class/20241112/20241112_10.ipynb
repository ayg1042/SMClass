```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
```

누적 막대 그래프

```python
plt.figure(figsize=(7, 4))
plt.bar(df['이름'], df['국어'], label='국어')
# bottom= 해당 값 위로 올라간다.
plt.bar(df['이름'], df['영어'], bottom=df['국어'], label='영어')
plt.bar(df['이름'], df['수학'], bottom=df['국어']+df['영어'], label='수학')
plt.xticks(rotation=45)
plt.legend(ncol=3, loc='upper center')
plt.ylim(00, 350)

# 값을 입력해서 넣어주세요
for i in range(len(df['국어'])):
  plt.text(df['이름'][i], df['국어'][i] + 0.5, df['국어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i] + 0.5, df['영어'][i], ha='center')
  plt.text(df['이름'][i], df['국어'][i] + df['영어'][i] + df['수학'][i] + 0.5, df['수학'][i], ha='center')

plt.show()
```
