```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
# 선형 그래프 생성
# plt.show()
# 타이틀
plt.figure(figsize=(7, 5))
plt.title('라인그래프 Line')
# 그래프 X축 설명, color 글자색 변경
plt.xlabel('X축 날짜', color='red', loc='left') # left, center, right
plt.ylabel('Y축 인원', color='blue', loc='top') # top, center, bottom
# plt.text(x위치, y위치, 데이터값), ha = 글자를 가운데 정렬
plt.text(x[0], y[0], 2)
# plt.text(x[1], y[1], 4)
for idx, txt in enumerate(y):
  plt.text(x[idx], y[idx], 4)
plt.plot(x, y, label='범례데이터', marker='o')
plt.legend()
plt.show()
```
