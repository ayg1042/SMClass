```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
yy = [1, 8, 3]
# 선형 그래프 생성
# linewidth = 선 두께, color : 선 색상
plt.plot(x, y, label='인원증가', linewidth=5, marker='o', markersize=10, markeredgecolor='red')
# plt.plot(x, yy, label='비용증가', marker='v', markersize=10, markerfacecolor='yellow')
# b 선색상, v 마커표시, : 선모양
plt.plot(x, yy, 'bv:', label='비용증가')
plt.title('라인그래프 Line', size=30)

# alpha : 두명도, ls : 라인 스타일 : 점선, -- dash선

# 범례 : 그래프 선 설명문구, loc : 범례 위치 upper, center, lower, left, right
# plt.legend(loc='upper right')
plt.legend(loc=(0.1, 0.8)) # 좌표값으로 표시
```

```python
## x,y축의 데이터
x = [1,2,3]  # 리스트,튜플,DataFrame데이터 - 리스트형태,Series
y = [2,4,8]
yy = [1,8,3]
## 선그래프 생성
# linewidth : 선두께,color : 선색상
# marker : 지점표시 - o,x,v, markersize : 마커크기,markeredgecolor: 마커테두리선
# markerfacecolor : 마커색상
# alpha : 투명도설정, linestyle:라인스타일 (:점선, -- dash선, -.)
plt.plot(x,y, label='인원변동',color='pink', linewidth=5, marker='o',markersize=10,markeredgecolor='red',markerfacecolor='yellow')
# plt.plot(x,yy, label='비용변동',marker='v',markersize=10,markerfacecolor='yellow',linestyle='--',alpha=0.2)
# plt.plot(x,yy, label='비용변동',marker='v',ms=20,mec='blue',mfc='green',ls='--',alpha=0.4)
###### 설정별 단축표시 'bv:'
# b:선색상,v:마커표시,: 선모양
plt.plot(x,yy,'bv:',label='비용변동')
## 상단타이틀
plt.title("라인그래프 Line",size=30)  # size : 글자크기 변경
## 범례 : 그래프 선 설명문구, loc:범례위치 upper,lower,left,center,right
# plt.legend(loc='upper center')
plt.legend(loc=(0.3,0.1))  # 좌표값으로 위치 설정 가능
```
