```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
# 선형 그래프 생성
plt.plot(x, y)
# plt.show()
# 타이틀
plt.title('라인그래프 Line')
# 그래프 X축 설명, color 글자색 변경
plt.xlabel('X축 날짜', color='red', loc='left') # left, center, right
plt.ylabel('Y축 인원', color='blue', loc='top') # top, center, bottom
# 그래프 단위
plt.xticks([1, 2, 3, 4, 5, 6, 7, 8])
plt.yticks([2, 4, 8])
plt.show()
```
