```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic' # win
mpl.rcParams['font.size'] = 15
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
```

```python
plt.plot(df['이름'], df['키'])
plt.xticks(rotation=45)
# 그래프 안내선, axis : x, y축으로 선이 그려진다.
# linestyle : 선 모양
# linewidth : 선 두께
# plt.grid(axis='y', linestyle='--', color='purple', linewidth=2)
# plt.grid(axis='x', linestyle='--', color='purple', linewidth=2)
plt.grid(axis='x')

# 그래프 값 추가
for i in range(8):
  plt.text(df['이름'][i], df['키'][i]+0.3, df['키'][i], ha='center')


plt.show()
```
