문자열 함수 : startwith, contains, replace, lower, upper


```python
# starwith : 해당값으로 시작하는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains : 해당값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt, ['이름', '학교', '키']]
# contains(na=False, True) : NaN데이터 처리해야한다.
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]

# isin : 포함이 되어 있는지 확인
# 데이터는 리스트 타입만 가능합니다.
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower() : 해당값을 모두 소문자로 변환
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 소문자로 변경후 j가 포함되어 있는 정보 검색
filt = df['SW특기'].str.lower().str.contains('j', na=False)
df[filt]
```

```python
## dropna
# 축 방향 axis = index, columns
# how = any(NaN값이 하나라도 있으면), all(모두 NaN값 이면)
# default - index, any
# df = pd.read_excel('score.xlsx', index_col='지원번호')
df2 = df.dropna().copy()
df2['학교'] = np.nan
df2.dropna(axis='index', how='all') # 모두 NaN인 row가 없어서 삭제 안됨
df2.dropna(axis='columns', how='all') # 학교 컬럼이 모두 NaN이기에 삭제됨
df2.dropna(axis='index', how='any') # 학교컬럼이 모두 nan 이기에 모든 row 삭제
df2.dropna(axis='columns', how='any') # 학교컬럼이 모두 nan 이기에 모든 row 삭제
```
