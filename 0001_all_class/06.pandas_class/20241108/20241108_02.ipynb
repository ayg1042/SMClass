데이터 추가

```python
# 데이터 추가
# 컬럼에 없는 컬럼 이름을 입력하면 컬럼이 추가 된다.
df['합계'] = 0
df['합계'] = df['국어'] + df['사회'] + df['영어'] + df['수학']
df['평균'] = df['합계']/4
df
```

```python
# df.drop(columns='결과', inplace=True)
df['결과'] = 'Fail'
filt = df['합계'] >= 300
df.loc[filt, '결과'] = 'Pass'
df
```

```python
df['합산'] = 100
## 국어점수가 90점 이상이면 df['합산'] + 100
filt = df['국어'] >= 90
df.loc[filt, '합산'] += 100
df
```
