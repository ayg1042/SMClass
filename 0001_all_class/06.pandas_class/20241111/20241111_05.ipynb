```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
# 관객수 * 10000 수정하고, 관객수로 역순하세요.
df
```

```python
df['관객 수'] = df['관객 수'] * 10000
```

```python
df.sort_values('관객 수')
```

```python
df['개봉 연도'] = df['개봉 연도'].astype(str) + '년도'
```

```python
df
```

```python
# 관객수 천단위 표시
df['관객 수'] = '$'+df['관객 수'].astype(str)
# df['관객 수'] = df['관객 수'].map('{:,d}'.format)
```

```python
df
```

```python
df['관객 수'] = df['관객 수'].str[1:].astype(int)
```

```python
# [2014, 2015, 2017, 2014, 2015, 2017, 2013, 2015]
df.groupby('개봉 연도')[['개봉 연도', '관객 수']].value_counts()
```
