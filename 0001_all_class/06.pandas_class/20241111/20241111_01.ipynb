```python
# str[:] : 문자 슬라이싱 substr과 같음
# str.split() : 분할
# str.startswith() : 시작글자 인식
# str.endswith() : 끝글자 인식
# str.contains() : 포함글자 인식
# str.find() : 문자위치찾기
# str.replace() : 문자 대체
# str.pad() : 무자열 패딩, 특정글자로 채워준다.
# str.zfill() : 0으로 채워줌
# str.strip() : 공백제거
# str.lower() : 소문자 변경
# str.upper() : 대문자 변경
# str.swapcase() : 대소문자 변경
```

```python
# 함수를 사용해서
# 이름 첫글자를 제거해서 이름을 입력하세요.
def names(name):
  return name[1:]
df['이름'] = df['이름'].apply(names)
df
```

```python
# 1. 요일 컬럼 추가
# 2. 0~6까지의 숫자를 랜덤으로 입력하세요.
import random as rd
# 3. 함수를 사용해서 0 - 일요일, 1 - 월요일, 2 - 화요일, 3 - 수요일 ...
df['요일'] = 0
df
for i in df.index:
  df.loc[i, '요일'] = rd.randrange(0, 7)
def today(day):
  days = ['일요일', '월요일', '화요일', '수요일', '목요일', '금요일', '토요일']
  return days[day]
df['요일'] = df['요일'].apply(today)
df
```
